```python
from operator import index


#constructor  __init__()
class employee:
    def __init__(self,first,last,salary):
        self.first = first                                            #initializing 'first' attribute to an object
        self.last = last                                              #initializing 'last' attribute to an object
        self.salary = salary                                          #initializing 'salary' attribute to an object
        self.email = f"{first.lower()}.{last.lower()}@company.com"    #creating email

emp1 = employee('Appoh','Augustine',50000)

print(emp1.first,emp1.last,emp1.salary)
print(emp1.email)
```

```python
#string representation method __str__()
class employee:
    def __init__(self,first,last,salary):
        self.first = first
        self.last = last
        self.salary = salary

    def __str__(self):
        return f"employee: {self.first}, {self.last}, {self.salary}"

emp = employee('Appoh','Augustine',50000)
print(emp)


#string representation  __repr__()
class employee:
    def __init__(self,first,last,salary):
        self.first = first
        self.last = last
        self.salary = salary

    def __repr__(self):
        return f"employee('{self.first}', '{self.last}', '{self.salary}')"

emp = employee('Mattew','Cobbinah',90000)
print(emp)
```

```python
#length method __len__()
class umat_halls:
    def __init__(self,halls):
        self.halls = halls

    def __len__(self):
        return len(self.halls)

umat_halls = umat_halls(['chambers of mines hall','gold refinery hall','kofi tetteh hall','rail hall'])
print(len(umat_halls))
```

```python
#equality method __eq__()
class languages:
    def __init__(self,languages):
        self.languages = languages

    def __eq__(self,other):
        return self.languages == other.languages

lan1 = languages('twi')
lan2 = languages('twi')

print(lan1 == lan2)
```

```python
#operator overloading
class vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __add__(self,other):
        return vector(self.x+other.x,self.y+other.y)
    def __sub__(self,other):
        return vector(self.x-other.x,self.y-other.y)
    def __mul__(self,other):
        return vector(self.x*other.x,self.y*other.y)
    def __truediv__(self,other):
        return vector(self.x/other.x,self.y/other.y)

    def __str__(self):
        return f'{self.x}i,{self.y}j'

v1 = vector(5,6)
v2 = vector(8,9)
print(v1 + v2)
print(v1 - v2)
print(v1*v2)
print(v1/v2)
```

```python
#item access __getitem__()
class cities:
    def __init__(self,cities):
        self.cities = cities

    def __getitem__(self, index):
        return self.cities[index]

cty = cities(['kumasi','tarkwa','obuasi','sunyani','accra'])
print(cty[3])
print(cty[2])
```

```python
class cars:
    def __init__(self,cars):
        self.cars = cars

    def __setitem__(self,index,value):
        self.cars[index] = value

car_obj = cars(['audi','benz','bmw','toyota'])
car_obj[1] = 'corolla'
print(car_obj.cars)
```

```python
#membership testing __contains__
class food:
    def __init__(self,food):
        self.food = food

    def __contains__(self, item):
        return item in self.food

food_obj = food(['akpele','banku','emutuo','fufu','jollof'])
print('cold rice' in food_obj)
print('banku' in food_obj)
print('emutuo' in food_obj)
```

```python
#callable object __call__
class  adder:
    def __init__(self,value):
        self.value = value

    def __call__(self, number):
        return self.value + number

add5 = adder(17)
print(add5(20))
```

```python
#iteration support  __iter__ and __next__
class EvenNumber:
    def __init__(self,limit):
        self.limit = limit
        self.current = 2

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration

        value = self.current
        self.current += 2
        return value

for num in EvenNumber(10):
    print(num)
```

```python
#boolean evaluation __bool__()
class SchoolFees:
    def __init__(self,balance):
        self.balance = balance

    def __bool__(self):
        return self.balance>0

account = SchoolFees(588)

if account:
    print("account is active")
```

```python
#destructor method __del__()
class resource:
    def __init__(self):
        print("resource allocated")

    def __del__(self):
        print("resource released")

r = resource()
del r
```
